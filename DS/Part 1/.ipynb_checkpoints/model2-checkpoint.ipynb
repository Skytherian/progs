We want to see where there is a strong linear effect or correlation on the target variable for creating a linear regression model, but first we will estimate the effects which are not linear in nature and yet influence the target variable.After getting the effects they will be included in our linear model. 
  
One such effect is the Number of convenience stores which greatly influences the target variable, we will use a Conditional Inference tree to get its effect on the target variable

### Splitting training and testing data

```r
set.seed(412)
train.index <- caret::createDataPartition(df$`Number of convenience stores`, p = .7, list = FALSE)
train <- df[ train.index,]
test  <- df[-train.index,]
```

```r
library(party, quietly = TRUE)
control <- ctree_control(mincriterion = 0.86, teststat = "quad",testtype = "Bonferroni")

tree <- ctree(`House price of unit area`~`Number of convenience stores`,controls = control, data=df) 
```

```r
plot(tree)
```

```r
library(party)


conve<-modeltools::Predict(tree,train[,4])

conve |> unique() |> head()
train$conve <- conve
```

### Decision trees
We create a decision tree to get the effect of house size and another decision tree to get the effect of number of bedrooms

```r
library(rpart)
treehsize<-rpart(`House price of unit area` ~`House size (sqft)`,data=train,method="anova")
print(treehsize)
plot(treehsize)
text(treehsize,xpd=TRUE)
treehsize$cptable
```

```r
treebedr <- rpart(`House price of unit area` ~ `Number of bedrooms`,data=train,cp=1e-5)

plot(treebedr)
text(treebedr,xpd=TRUE)
```

```r
treemonth <- rpart(`House price of unit area` ~ Transaction_Month,data=train,method="anova",cp=0.009)
print(treemonth)

plot(treemonth)
text(treemonth, xpd =TRUE)
```

```r
hsize<-predict(treehsize,train[,c('House price of unit area','House size (sqft)')])
train$hsize <- hsize
bedr <- predict(treebedr, train[,c('House price of unit area', 'Number of bedrooms')])
train$bedr <- bedr
month <- predict(treemonth, train[,c('House price of unit area', 'Transaction_Month')])
train$month <- month
```

```r
train |> select(where(is.numeric)) |> cor()
```

We will be including interaction effects in the model, the reasons are:  
Areas near metro station should have more convenience stores  
Number of bedrooms would be more in houses with greater house size

```r
m1 <- lm( `House price of unit area` ~ `House Age`+ `Distance from nearest Metro station (km)`:conve+ `Transaction date`:month +latitude:longitude+I(conve^2)+hsize:bedr,data = train)       
summary(m1)

boxinvTransform <- function(y, lambda) {
  if (lambda == 0L) { exp(y) }
  else { (y * lambda + 1)^(1/lambda) }
}
```

```r
check_vif <- function(model)
{
    
resid1 <- residuals(model)
library(car,quietly=TRUE)
vif(model) |> print()

vif_values <- vif(model)           #create vector of VIF values

barplot(vif_values, main = "VIF Values", horiz = TRUE, col = "steelblue") #create horizontal bar chart to display each VIF value

abline(v = 5, lwd = 3, lty = 2)    #add vertical line at 5 as after 5 there is severe correlation
}
check_vif(m1)
```

#### We create a WLS model and then do a Box Cox transformation

```r
wt <- 1 / lm(abs(m1$residuals) ~ m1$fitted.values)$fitted.values^2
wls_model <- lm( `House price of unit area` ~ `House Age`+ `Distance from nearest Metro station (km)`:conve+ `Transaction date`:month +latitude:longitude+I(conve^2)+hsize:bedr,data = train,weights=wt)       
summary(wls_model)
bc<-MASS::boxcox(wls_model)
(lambda <- bc$x[which.max(bc$y)])
```

```r
m1 <- lm( ((`House price of unit area`^lambda-1)/lambda) ~`House Age`+ `Distance from nearest Metro station (km)`:conve+ `Transaction date`:month +latitude:longitude+I(conve^2)+hsize:bedr,data = train,weights=wt)
summary(m1)
performance::check_model(m1)
```

We define mean absolute percentage error as  

Mean of ( (|Predicted - Actual|)/Actual * 100 )

```r
percent_error <- function(resid,actual,title)
    {
    
    percent <- resid/actual*100
    
    percent |> Filter(\(x) x> -100 & x < 100, x=_) |> hist(xlab = "Percent deviation", main = title)
    
    percent |> abs() |> mean() -> error 
    
    library(glue)
    mtext(glue("The mean absolute {title} in percentage is { round(error,digits=3) }"), side =3)
    
    return (error)
    
    
    }
resid1 <- boxinvTransform(predict(m1,newdata=train),lambda ) - train$`House price of unit area`

rmse <- function(resid,title){
    
    resid^2 |> mean() |> sqrt() -> rms
    hist(resid, xlab= "Residual values")
    
    library(glue)
    mtext(glue("The Root mean square error of {title}  is { round(rms,digits=3) }"), side =3)
    return (rms)
}
```

```r
rmse(resid1, "Training set")
```

```r
percent_error(resid1, train$`House price of unit area`, "Training deviation") 
```
